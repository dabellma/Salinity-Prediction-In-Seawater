```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('epochs_vs_loss.csv')

plt.figure(figsize=(8, 5))
plt.plot(data['Epoch'], data['MSE Loss'], marker='o')
plt.title('Epoch vs. MSE Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(data['Epoch'], data['MSE Loss'], marker='o')
plt.yscale('log')
plt.title('Epoch vs. Log MSE Loss')
plt.xlabel('Epoch')
plt.ylabel('Log MSE Loss')
plt.grid(True)
plt.show()
```
